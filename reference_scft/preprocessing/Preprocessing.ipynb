```python
#remove background

BLUR = 21
CANNY_THRESH_1 = 15
CANNY_THRESH_2 = 100
MASK_DILATE_ITER = 20
MASK_ERODE_ITER = 20
MIN_AREA = 0.0005
MAX_AREA = 0.2
MASK_COLOR = (1.0, 1.0, 1.0)

def removeBackground(image):
    # Convert image to grayscale 
    frame = cv2.imread(image)
    image_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Apply Canny Edge Dection
    edges = cv2.Canny(image_gray, CANNY_THRESH_1, CANNY_THRESH_2)
    edges = cv2.dilate(edges, None)
    edges = cv2.erode(edges, None)
     # get the contours and their areas
    contour_info = [(c, cv2.contourArea(c),) for c in cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)[0]]

    # Get the area of the image as a comparison
    image_area = frame.shape[0] * frame.shape[1]  

    # calculate max and min areas in terms of pixels
    max_area = MAX_AREA * image_area
    min_area = MIN_AREA * image_area
    # Set up mask with a matrix of 0's
    mask = np.zeros(edges.shape, dtype = np.uint8)
    
    # Go through and find relevant contours and apply to mask
    for contour in contour_info:
        # Instead of worrying about all the smaller contours, if the area is smaller than the min, the loop will break
        if contour[1] > min_area and contour[1] < max_area:
            # Add contour to mask
            mask = cv2.fillConvexPoly(mask, contour[0], (255))
    # use dilate, erode, and blur to smooth out the mask
    mask = cv2.dilate(mask, None, iterations=MASK_DILATE_ITER)
    mask = cv2.erode(mask, None, iterations=MASK_ERODE_ITER)
    mask = cv2.GaussianBlur(mask, (BLUR, BLUR), 0)
    
    mask_stack = np.dstack([mask] * 3)
    # Ensures data types match up
    mask_stack = mask_stack.astype('float32') / 255.0           
    frame = frame.astype('float32') / 255.0

    # Blend the image and the mask
    masked = (mask_stack * frame) + ((1-mask_stack) * MASK_COLOR)
    masked = (masked * 255).astype('uint8')
    cv2.imwrite("backgroundclean.png", masked)

# removeBackground("wp7823331.png") 
```

```python
import cv2
import numpy as np
import glob
    
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)

def ColorToPencilLineArt(directoryPath):
    imgList = glob.glob('%s/*.png'% directoryPath)
    count = 0
    for path in imgList[:1]:
        img = cv2.imread(path, cv2.IMREAD_COLOR)
        
        #convert to greyscale
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        #make brighter regions even brighter to make it easier to detect edges
        img_invert = cv2.bitwise_not(img_gray)
        
        #smooth image
        img_smoothing = cv2.GaussianBlur(img_invert, (45, 45), sigmaX=0, sigmaY=0)
        
        #highlight the boldest portions of the 'lineart'
        final_img = dodgeV2(img_gray, img_smoothing)

        #use OTSU thresholding + "close" morphology transform
        #^ FAIL
        
        success = cv2.imwrite("../data/linearts/"+ path.rsplit('/', 1)[-1], final_img)     # save frame as JPEG file      
        #success = cv2.imwrite("../data/linearts/"+ str(test) +".png", final_img)     # save frame as JPEG file      

        count += 1
    print(count, " images converted to lineart")

ColorToPencilLineArt('../data/midstep')
```
